```python
num1 = float(input("Enter first number: "))
num2 = float(input("Enter second number: "))
opt = input("Enter the operation: ")
if opt == "+":
    print(num1 + num2)
elif opt == "-":
    print(num1 - num2)
elif opt == "*":
    print(num1 * num2)
elif opt == "/":
    if num2 == 0:
        print("Error: Division by zero")
    else:
        print(num1 / num2)
else:
    print("Invalid operation")
```
